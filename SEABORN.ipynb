```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Set_theme

```python
sns.barplot(x=["A", "B", "C"], y=[1,3,2])
plt.show()
```

# Distribuciones
### Histograma
#### bins = cantidad de barras para la distribucion
#### cumulative = acumula el valor progresivamente de cada barra
#### stat = define la forma de la distribucion (frequency, probability, percent, density, count )
#### multiple = como presentar los datos (stack, layer, dodge, fill)

```python
#Histograma
sns.histplot(data=tips, x="tip", bins=15, cumulative=False, hue="sex", stat="count", multiple="dodge")
plt.show()
```

### Diagrama de densidad
#### shade = muestra area bajo la curva
#### bw_adjust = ajusta la grafica y los parametros de los ejes

```python
#Diagrama de densidad del comportamiento de propinas segun el sexo
sns.kdeplot(data=tips, x="tip", hue="sex", cumulative=False, shade=True, bw_adjust=1)
plt.show()
```

### ecdf plot

```python
# Grafico escalonado de la distribucion que tienen las propinas
sns.ecdfplot(data=tips, x="tip", hue="sex", stat="count")
plt.show()
```

### Displot

```python
# displot es la herramienta general que se puede usar para cualquier distribucion cambiando kind
sns.displot(data=tips, x="tip", hue="sex", kind="hist", multiple="stack")
plt.show()
```

# Categoricos

### Countplot

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
```

### Stripplot
#### dodge = separa la distribucion de acuerdo al hue

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

### Swarmplot
#### Mejor forma de ver la distribucion donde se agrupan mucho los puntos

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

### Mezclar dos graficas (boxplot - swarmplot)

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, color="0")
plt.show()
```

### Violinplot
#### split = combina el hue que este usando

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.show()
```

### Catplot
#### kind = define el tipo de grafico categorico a utilizar
#### col = agregar una variable adicional y partirla en columnas

```python
#catplot es la herramienta general con la que puedo usar cualquier grafico para categoricas
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, kind="box", col="time")
plt.show()
```

# Relation
### Scatterplot

```python
#Se puede ajustar la posicion de la leyenda como con matplotlib
plt.figure(figsize=(8,8))
sns.scatterplot(data=tips, x="total_bill", y="tip", palette="dark", hue="day", style="time", size ="size")
plt.legend(loc="center", bbox_to_anchor=(1.12,0.5))
plt.show()
```

### Lineplot

```python
plt.figure(figsize=(8,8))
sns.lineplot(data=tips, x="total_bill", y="tip", palette="dark", hue="day", style="time", size ="size")
plt.legend(loc="center", bbox_to_anchor=(1.12,0.5))
plt.show()
```

### Relplot

```python
#relplot es la herramienta general con la que puedo usar cualquier grafico de dispersion
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", palette="dark", col="time")
plt.show()
```

# Joinplot / Pairplot
#### kind = scatter, hist, hex, kde, reg, resid
#### marginal_ticks = me agrega escala a la segunda grafica
#### marginal kws = me permite  configurar a detalle la segunda grafica
Une dos distintas graficas en una sola para mejorar los analisis

```python
sns.jointplot(data=tips, x="total_bill", y="tip", hue="sex", kind="hist", marginal_ticks=True,
             marginal_kws=dict(bins=25, fill=True, multiple="dodge"))
plt.show()
```

```python
# paiplot me muestra la distribucion entre las variables numericas
sns.pairplot(data=tips, hue="sex")
plt.show()
```

# Heatmap
#### annot = agrega valores de correlacion en el grafico
#### linewidths = separa los cuadrantes

```python
# Correlacion entre variables -1 = totalmente indirecta, 0 = no hay correlacion, 1 = correlacion directa
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot=True, linewidths=1)
plt.show()
```
